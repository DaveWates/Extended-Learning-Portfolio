```python
def ave_Yrly_RF():
 df_list = ['Jan', 'Feb', 'Mar','Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] 
 df['Ave Yearly Rain'] = df[df_list].mean(axis=1)
 return df
```

```python
def min_RF():
 df_list = ['Jan', 'Feb', 'Mar','Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] 
 df['Minimum Yearly Rain'] = df[df_list].min(axis=1)
 return df['Minimum Yearly Rain']
```

```python
def max_RF():
 df_list = ['Jan', 'Feb', 'Mar','Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] 
 df['Maximum Yearly Rain'] = df[df_list].max(axis=1)
 return df['Maximum Yearly Rain']
```

```python
rainfall = pd.read_csv('rainfall.csv')

plt.figure(figsize= (10,6))
plt.title('Rainfall Over Time')
plt.plot(rainfall.Year, rainfall.Jan, label = 'Jan')
plt.plot(rainfall.Year, rainfall.Feb,label = 'Feb')
plt.plot(rainfall.Year, rainfall.Mar,label = 'Mar')
plt.plot(rainfall.Year, rainfall.Mar,label = 'Annual')
plt.legend()


```
